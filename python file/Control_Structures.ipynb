# Iterating through a range

```python
r = range(10,51,5)
```

```python
for i in r:
    print(i)
```

```python
lis = [1,2,5,6,8]
```

```python
for item in lis:
    x = item**2
    print(x)
```

```python
websites = ["amazon", "flipkart", "snapdeal"]
```

```python
new_webs = []
for sites in websites:
    #print("www."+sites+".com")
    new_webs.append("www."+sites+".com")
```

## Iteration through dict

```python
vk = {"name":"Virat", "Age":33, "Team": "RCB"}
```

```python
vk.keys()
```

```python
vk.values()
```

```python
for i in vk:
    print(i)
```

```python
for i in vk.values():
    print(i)
```

```python
for i in vk.items():
    print(i)
```

## Break

```python
for i in range(1,11):
    print(i)
    if i == 5:
        break
```

```python
## Continue
for i in range(1,11):
    if i%2 != 0:
        continue
    print(i)
```

```python
for i in range(1,11):
```

```python
for i in range(1,11):
    pass
```

```python
# Write a program to get sum of all elements in a list (dont use inbuilt functions)
# input = [1,3,5,1]
# Output = 10
```

```python
num = [1,3,5,1]
res = 0
for i in num:
    res = res+i
print("sum of list:", res)
```
